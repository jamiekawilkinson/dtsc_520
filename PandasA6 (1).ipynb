```python
import numpy as np
import pandas as pd
```

```python
dotsdf = pd.read_csv("dots.csv")
```

```python
# Obtain the mean of the final column and assign it to Q1.

Q1 = dotsdf['firing_rate'].mean()
```

```python
# Obtain the maximum time and assign it to Q2.

Q2 = dotsdf['time'].max()
```

```python
# Use the groupby method to obtain the mean values of each column for T1 and T2.  Assign it to Q3.

Q3 = dotsdf.groupby('choice').mean()
```

```python
# Use the groupby method to obtain the minimum firing rates for dots and sacc.  It should look like this:

Q4 = dotsdf.groupby('align')['firing_rate'].min()
```

```python
# Use the groupby method, grouping by the first two columns,to obtain the mean firing rate values.

q5 = dotsdf.groupby(['align','choice'])['firing_rate'].mean()
Q5 = q5.unstack()
```

```python
# Use a pivot table to obtain the same as Q5.  Assign it to Q6.

Q6 = dotsdf.pivot_table('firing_rate', index='align', columns='choice')
```

```python
tipsdf = pd.read_csv('tips.csv')
```
