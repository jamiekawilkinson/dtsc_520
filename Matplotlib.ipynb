```python
import numpy as np
import matplotlib as mlt
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]

ax.plot(x,y)

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

```python
#change size of plot

fig, ax = plt.subplots(figsize = (3,5))

x = [1,2,3,4]
y = [3,4,2,1]

ax.plot(x,y)

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

```python
#change line and x/y limits

fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]

ax.plot(x,y, 'o')

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.xlim(0,5)
plt.ylim(0,5)

plt.show()
```

```python
#changle line style

fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]

ax.plot(x,y, '--')

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

```python
#change color

fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]

ax.plot(x,y, 'r')

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

```python
#change multiple styles at once

fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]

ax.plot(x,y, '--r+')

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

```python
#change size of the line

fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]

ax.plot(x,y,linewidth = 20)

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

```python
#change markers

fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]

ax.plot(x,y, 'o', markerfacecolor='aqua', markeredgecolor='orchid',
       markeredgewidth=4, markersize=20)

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

```python
#remove grid lines

fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]

ax.plot(x,y, 'tan')

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.grid(False)

plt.show()
```

```python
#change grid color
fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]

ax.plot(x,y)

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.grid(color = 'brown')

plt.show()
```

```python
#multiple lines
#c means color
#ls means linestyle
fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]
a = [.5, 1.5, 2.5, 3.5]
b = [3.5, 1.5, .5, 2.5]

ax.plot(x,y,'purple')
ax.plot(a,b, c = 'brown', ls='-.')

plt.title("My figure")

plt.xlim(0,5)
plt.ylim(0,5)

plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
#change tickmarks and add labels

fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]
a = [.5, 1.5, 2.5, 3.5]
b = [3.5, 1.5, .5, 2.5]

ax.plot(x,y,'purple')
ax.plot(a,b, c = 'brown', ls='-.')

plt.title("My figure")

plt.xlim(0,5)
plt.ylim(0,5)

plt.xlabel("X")
plt.ylabel("Y")

plt.xticks([1,2,3])
plt.yticks([1,4], ['Sad', 'Happy'])

plt.show()
```

```python
#add legend

fig, ax = plt.subplots()

x = [1,2,3,4]
y = [3,4,2,1]
a = [.5, 1.5, 2.5, 3.5]
b = [3.5, 1.5, .5, 2.5]

ax.plot(x,y,'purple')
ax.plot(a,b, c = 'brown', ls='-.')

plt.title("My figure")

plt.xlim(0,5)
plt.ylim(0,5)

plt.xlabel("X")
plt.ylabel("Y")

plt.xticks([1,2,3])
plt.yticks([1,4], ['Sad', 'Happy'])

ax.legend(['Line 1', 'Line 2'])

plt.show()
```
